````python
import re
import os
import glob
import zipfile
import requests
from urllib.request import urlretrieve
import json
import pandas as pd
import boto3
import awswrangler as wr
````

Rememeber here we gave the folder that we created in Step 3.2 as we made it available for all the users in a group.

````python
# Necessary metadata
article_id = 14226968  # this is the unique identifier of the article on figshare
url = f"https://api.figshare.com/v2/articles/{article_id}"
headers = {"Content-Type": "application/json"}
output_directory = "/srv/data/my_shared_data_folder/"
````

````python
response = requests.request("GET", url, headers=headers)
data = json.loads(response.text)  # this contains all the articles data, feel free to check it out
files = data["files"]             # this is just the data about the files, which is what we want
files
````

````python
files_to_dl = ["combined_model_data_parti.parquet.zip"]  ## Please download the partitioned 
for file in files:
    if file["name"] in files_to_dl:
        os.makedirs(output_directory, exist_ok=True)
        urlretrieve(file["download_url"], output_directory + file["name"])
````

````python
with zipfile.ZipFile(os.path.join(output_directory, "combined_model_data_parti.parquet.zip"), 'r') as f:
    f.extractall(output_directory)
````

````python
# Filter for Sydney
df_syd = df.query('(lon_min<151.21<lon_max) & (lat_min<-33.86<lat_max)').drop(['lat_min', 'lat_max', 'lon_min', 'lon_max'], axis=1)
````

````python
# Format the datetime
df_syd['time'] = pd.DatetimeIndex(df_syd['time']).strftime("%Y-%m-%d")
````

````python
# Add observed column
obs_df['model'] = 'Observed'
obs_df['time'] = pd.DatetimeIndex(obs_df['time']).strftime("%Y-%m-%d")

# Add observed data to models df
df_syd = pd.concat([df_syd, obs_df])
````

````python
# Pivot the table wide
syd_wide = pd.pivot_table(df_syd, values="rain (mm/day)", index=["time"], columns=["model"])
````
